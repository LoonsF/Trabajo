```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

imagen = cv2.imread('Gat.jpg')
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
filtros_bayer = [cv2.COLOR_BAYER_BG2BGR, cv2.COLOR_BAYER_GB2BGR, cv2.COLOR_BAYER_RG2BGR, cv2.COLOR_BAYER_GR2BGR]
nombres_filtros_bayer = ['BAYER_BG2BGR', 'BAYER_GB2BGR', 'BAYER_RG2BGR', 'BAYER_GR2BGR']

plt.figure(figsize=(10, 8))
for i, filtro_bayer in enumerate(filtros_bayer):
    imagen_bayer = cv2.cvtColor(imagen_gris, filtro_bayer)
    plt.subplot(2, 2, i + 1)
    plt.imshow(cv2.cvtColor(imagen_bayer, cv2.COLOR_BGR2RGB))
    plt.title(nombres_filtros_bayer[i])
    plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

imagen = cv2.imread('Gat.jpg')
imagen_yuv = cv2.cvtColor(imagen, cv2.COLOR_BGR2YUV)
rgb_desde_yuv = cv2.cvtColor(imagen_yuv, cv2.COLOR_YUV2RGB)
yuv_desde_rgb = cv2.cvtColor(rgb_desde_yuv, cv2.COLOR_RGB2YUV)

plt.figure(figsize=(10, 8))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(imagen_yuv, cv2.COLOR_YUV2RGB))
plt.title('Imagen YUV')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(rgb_desde_yuv, cv2.COLOR_RGB2BGR))
plt.title('RGB desde YUV')
plt.axis('off')

plt.show()
```
